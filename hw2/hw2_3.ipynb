```python
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
```

```python
def adaptive(grayscaled, blocksize=15, param=5):
    return cv2.adaptiveThreshold(grayscaled, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, blocksize, param)


def blur_threshold(grayscaled, th, k_size=5):
    image1 = cv2.GaussianBlur(grayscaled,(k_size,k_size),0) 
    ret, image2 = cv2.threshold(image1, th, 255, cv2.THRESH_BINARY)
    return image2

def otsu(grayscaled):
    retval, im = cv2.threshold(grayscaled, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    return im

```

## Image 1

```python
img = cv2.cvtColor(io.imread("tasks/object/obj1.jpg"), cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
(w,h) = img.shape[:2]
mask = np.zeros((w,h),np.uint8)
bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
rect = (25,25,430,210)
mask, bgdModel, fgdModel = cv2.grabCut(img,mask,rect,bgdModel,fgdModel,15,cv2.GC_INIT_WITH_RECT)
mask = np.where((mask==2)|(mask==0),0,1).astype('uint8')
image_segmented = img*mask[:,:,np.newaxis]
```

```python
plt.figure(figsize=(24, 16))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
```

## Image 6 (Color of the object is very similar to background and we need to do a lot of manual marking to get at least not terrible result)

```python
img = cv2.cvtColor(io.imread("tasks/object/obj6.jpg"), cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
(w,h) = img.shape[:2]
mask = np.zeros((w,h),np.uint8)
mask[:, :] = 2
mask[150:350, 100:200] = 1
mask[350:400, 100:120] = 1
mask[350:400, 175:225] = 1
mask[350:400, 0:90] = 0
mask[460:500, 0:300] = 0
mask[:, 250:] = 0

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)
mask, bgdModel, fgdModel = cv2.grabCut(img,mask,None,bgdModel,fgdModel,5,cv2.GC_INIT_WITH_MASK)
mask = np.where((mask==2)|(mask==0),0,1).astype('uint8')
image_segmented = img*mask[:,:,np.newaxis]
```

```python
plt.figure(figsize=(24, 16))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(image_segmented, cv2.COLOR_BGR2RGB))
```
